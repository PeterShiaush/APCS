## **【第一節：如果...就...的選擇邏輯】 (50 分鐘)**

### **1. [教學區]：關係運算子與縮排**

在 Python 中，我們使用以下符號來比較兩邊的大小：

* `==`：等於（**注意**：一個 `=` 是賦值，**兩個** `==` 才是比較）。
* `!=`：不等於。
* `>`, `<`：大於、小於。
* `>=`, `<=`：大於等於、小於等於。

**語法結構：**

```
if 條件式:
    # 這裡必須「縮排」(按 Tab 鍵)，代表條件成立才執行
    執行指令
else:
    # 當條件不成立時執行
    執行指令
```

* **閏年規則**：四年一閏，百年不閏，四百年又一閏。
* **提示**：這題需要處理「多筆輸入」，請直接套用下方的 `while` 迴圈框架。

```
import sys
# 這是處理多筆輸入的標準寫法
for line in sys.stdin:
    year = int(line)
    # 邏輯：(能被4整除 且 不能被100整除) 或 (能被400整除)
    if (year % 4 == 0 and year % 100 ??? 0) or (year % 400 ??? 0):
        print("閏年")
    else:
        print("平年")
```
